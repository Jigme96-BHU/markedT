```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
```

```python
california = fetch_california_housing()
X,y = california.data, california.target
feature_names = california.feature_names


```

```python
feature_names
```

```python
#Splitting Data
#70% for train, 20% test, 20% val
x_train_validate, x_test, y_train_validate, y_test = train_test_split(X,y, test_size=0.2,random_state=33)
x_train, x_val,y_train,y_val = train_test_split(x_train_validate,y_train_validate,test_size=0.2,random_state=33)
```

```python
#data standardisation 

# to have standard range for all the feature(mean = 0 and SD= 1)
# of we dont do standardisatioin, if a feature has a variance that is much larger than other feature
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(x_train)
x_train_scaler = scaler.transform(x_train)
```

```python
#Feed forward neural network 
#we use a neural network to learn the relationship between the target and the feature
from sklearn.neural_network import MLPRegressor

mlp_regressor = MLPRegressor(verbose=True).fit(x_train_scaler,y_train)
```

```python
from sklearn.neural_network import MLPRegressor

mlp_regressor = MLPRegressor(verbose= True).fit(x_train_scaler, y_train)

print("Training score of MLP:", mlp_regressor.score(x_train_scaler,y_train))

# also standardise test data
x_val_scaled = scaler.transform(x_val)
print("Validation score of MLP:", mlp_regressor.score(x_val_scaled, y_val))

# also standardise test data
x_test_scaled = scaler.transform(x_test)
print("accuracy on unseen  of MLP:", mlp_regressor.score(x_test_scaled, y_test))
```

```python
#fine tunning the network 

mlp_regressor = MLPRegressor(verbose=True, hidden_layer_sizes=(50,),
                            activation='relu',
                            solver='adam',
                            batch_size=20,
                            learning_rate_init=0.001).fit(x_train_scaler,y_train)

print('Training socre of MLP:', mlp_regressor.score(x_train_scaler, y_train))

x_val_scaled = scaler.transform(x_val)
print(' validation socre of MLP:', mlp_regressor.score(x_val_scaled, y_val))

x_test_scaled = scaler.transform(x_test)
print('accarcuy of unseen data socre of MLP', mlp_regressor.score(x_test_scaled, y_test))

```

```python
#Load Dataset

from sklearn.datasets import load_iris

```

```python
irish_data = load_iris()
```

```python
X,y = irish_data.data, irish_data.target
feature_name = irish_data.feature_names
```

```python
#Splitting Data
#70% for train, 20% test, 20% val
x_train_validate, x_test, y_train_validate, y_test = train_test_split(X,y, test_size=0.3,random_state=33)
x_train, x_val,y_train,y_val = train_test_split(x_train_validate,y_train_validate,test_size=0.1,random_state=33)
```

```python
#data standardisation 

# to have standard range for all the feature(mean = 0 and SD= 1)
# of we dont do standardisatioin, if a feature has a variance that is much larger than other feature
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(x_train)
x_train_scaler = scaler.transform(x_train)
```

```python
from sklearn.neural_network import MLPClassifier

mlp_classifier = MLPClassifier(verbose= True).fit(x_train_scaler, y_train)

print("Training score of MLP:", mlp_classifier.score(x_train_scaler,y_train))

# also standardise test data
x_val_scaled = scaler.transform(x_val)
print("Validation score of MLP:", mlp_classifier.score(x_val_scaled, y_val))

# also standardise test data
x_test_scaled = scaler.transform(x_test)
print("accuracy on unseen  of MLP:", mlp_classifier.score(x_test_scaled, y_test))
```

```python

mlp_classifier = MLPClassifier(verbose=True, hidden_layer_sizes=(50,),
                            activation='relu',
                            solver='adam',
                            batch_size=20,
                            learning_rate_init=0.001).fit(x_train_scaler,y_train)

print('Training socre of MLP:', mlp_classifier.score(x_train_scaler, y_train))

x_val_scaled = scaler.transform(x_val)
print(' validation socre of MLP:', mlp_classifier.score(x_val_scaled, y_val))

x_test_scaled = scaler.transform(x_test)
print('accarcuy of unseen data socre of MLP', mlp_classifier.score(x_test_scaled, y_test))

```

```python
from sklearn.metrics import confusion_matrix

y_pred = mlp_classifier.predict(x_test_scaled)
cm = confusion_matrix(y_test, y_pred)
```
